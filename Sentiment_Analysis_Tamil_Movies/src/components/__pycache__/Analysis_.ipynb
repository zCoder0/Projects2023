```python
import json
import re
from langdetect import detect
import emoji
import pandas as pd
```

```python
# Function to split tweet components
def split_tweet(tweet):

    if isinstance(tweet['text'], list):
        text = ' '.join(tweet['text'])
    else:
        text = tweet['text']
    # Remove \n and \r
    text = text.replace('\n', ' ').replace('\r', ' ')
    #remove nuremeric

    text = re.sub(r'\d+', '', text)
    # Remove URLs
    urls = re.findall(r'http\S+', text)
    text = re.sub(r'http\S+', '', text)

    hashtags = re.findall(r'#\w+', text)
    text = re.sub(r'#\w+', '', text)
    emojis = ''.join(c for c in text if c in emoji.EMOJI_DATA)
    text = re.sub(r'[^\w\s]', '', text)
    return {
        'text': text.strip(),
        'urls': urls if urls else None,
        'hashtags': hashtags if hashtags else None,
        'emojis': emojis if emojis else None
    }
```

```python
tweet_data = [split_tweet(tweet) for tweet in tweets]
```

```python
def split_languages(text):
    words = text.split()
    print(words)
    print()
    tamil = ' '.join(w for w in words if  detect(w) == 'ta')
    english = ' '.join(w for w in words if detect(w) != 'ta')
    return pd.Series([tamil if tamil else None, english if english else None], index=['tamil', 'english'])
```

```python

# Load DataFrame (assuming it's already created)
# If not, run previous code to get df

# Initialize translator
translator = Translator()

# Sentiment analysis function
def get_sentiment(row):
    tamil_text = str(row['tamil'])
    english_text = str(row['english'])
    emojis = str( row['emojis'])

    print(tamil_text)
    print(english_text)
    print(emojis)

    print(type(tamil_text))
    print(type(english_text))
    print(type(emojis))

    # Tamil sentiment (translate to English)
    tamil_score = 0
    if tamil_text:
        try:
            translated = translator.translate(tamil_text, dest='en').text
            tamil_score = TextBlob(translated).sentiment.polarity
        except:
            tamil_score = 0  # Fallback if translation fails
    print("hi")

    print('before')
    # English sentiment
    english_score = 0
    if english_text:
      print("hello")
      english_score = TextBlob(english_text).sentiment.polarity

    # Emoji sentiment
    emoji_score = 0
    if emojis:
          emoji_text = emoji.demojize(emojis, delimiters=(' ', ' '))
          emoji_score = TextBlob(emoji_text).sentiment.polarity
    # Combine scores (weighted average)
    final_score = (tamil_score * 0.45) + (english_score * 0.45) + (emoji_score * 0.15)
    sentiment = 'Positive' if final_score > 0 else 'Negative' if final_score < 0 else 'Neutral'
    return pd.Series([final_score, sentiment], index=['score', 'sentiment'])
```

```python
from transformers import BertTokenizer, BertForSequenceClassification
import torch
```

```python
# Load FineBERT model and tokenizer
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForSequenceClassification.from_pretrained(model_name)

# Initialize translator
translator = Translator()
```

```python
# Sentiment function with FineBERT
def get_finebert_sentiment(row):
    tamil_text = str(row['tamil'])
    english_text = str(row['english'])

    # Combine text (translate Tamil to English)
    full_text = ""
    print(tamil_text)
    if tamil_text  !="nan":
        print("no")
        try:
            translated = translator.translate(tamil_text, dest='en').text
            full_text += translated + " "
        except:
            full_text += ""  # Skip if translation fails
    if english_text !="nan" :
        print("yes")
        full_text += english_text
    print(full_text)

    if not full_text.strip():
        print("empty")
        return pd.Series([0, "Neutral"], index=['score', 'sentiment'])

    # Tokenize and predict
    inputs = tokenizer(full_text, return_tensors="pt", truncation=True, padding=True, max_length=128)
    print(inputs)
    with torch.no_grad():
        outputs = model(**inputs)
        scores = outputs.logits.softmax(dim=-1).numpy()[0]  # [negative, neutral, positive]

    sc =[]
    sc.append(scores[1])
    sc.append(scores[2])
    sc.append(scores[3])
    print(sc)

    # Map scores
    print('scores ',scores)
    sentiment_map = {0: "Negative", 1: "Neutral", 2: "Positive"}
    score = sc[2] - sc[0]  # Positive - Negative (range -1 to 1)
    print(score)
    sentiment = sentiment_map[sc.index(max(sc))]
    print(sentiment)
    return pd.Series([score, sentiment], index=['score', 'sentiment'])
```

```python
# Apply FineBERT sentiment
df_copy2[['finebert_score', 'finebert_sentiment']] = df_copy2.apply(get_finebert_sentiment, axis=1)
print("FineBERT sentiment analysis done da!")
```

```python

import pandas as pd
from transformers import BertTokenizer, BertForSequenceClassification
from googletrans import Translator
import torch
import emoji


# Translator
translator = Translator()

# Expanded emoji sentiment dictionary
emoji_sentiment = {
    "üòä": 0.2, "üëç": 0.2, "üòÇ": 0.15, "üî•": 0.15, "üéâ": 0.2, "‚ù§": 0.2, "üòç": 0.2, "‚ú®": 0.15, "üìà": 0.15,  # Positive
    "üòû": -0.2, "üëé": -0.2, "üò°": -0.15, "üíî": -0.15, "üò¢": -0.2, "üò†": -0.15,                     # Negative
    "üòê": 0.0, "ü§î": 0.0, "ü§∑": 0.0, "üëÄ": 0.0, "üôÑ": 0.0                                     # Neutral
}

# Dynamic emoji sentiment function
def get_emoji_boost(emojis):
    if not emojis:
        return 0
    boost = 0
    for e in emojis:
        if e in emoji_sentiment:
            boost += emoji_sentiment[e]
        else:
            # Dynamic check using emoji description
            desc = emoji.demojize(e, delimiters=("", "")).lower()
            if any(word in desc for word in ["happy", "love", "fire", "up", "heart", "good"]):
                boost += 0.1  # Positive
            elif any(word in desc for word in ["sad", "angry", "broken", "down", "bad"]):
                boost -= 0.1  # Negative
            else:
                boost += 0.0  # Neutral
    return boost

# Sentiment function with emoji boost
def get_finebert_sentiment(row):
    tamil_text = str(row['tamil'])
    english_text = str(row['english'])
    emojis = str(row['emojis'])

    full_text = ""
    if tamil_text !="nan":
        try:
            full_text += translator.translate(tamil_text, dest='en').text + " "
        except:
            full_text += ""
    if english_text != "nan":
        full_text += english_text

    if not full_text.strip():
        return pd.Series([0, "Neutral"], index=['score', 'sentiment'])

    # FineBERT prediction
    inputs = tokenizer(full_text, return_tensors="pt", truncation=True, padding=True, max_length=128)
    with torch.no_grad():
        outputs = model(**inputs)
        scores = outputs.logits.softmax(dim=-1).numpy()[0]

    # Base score
    base_score = scores[2] - scores[0]

    # Emoji boost
    emoji_boost = get_emoji_boost(emojis)

    # Final score
    final_score = base_score + emoji_boost

    # Sentiment with threshold
    if final_score > 0.1:
        sentiment = "Positive"
    elif final_score < -0.1:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"

    return pd.Series([final_score, sentiment], index=['score', 'sentiment'])

# Apply to DataFrame
df_copy2[['finebert_score', 'finebert_sentiment']] = df_copy2.apply(get_finebert_sentiment, axis=1)
print("FineBERT with dynamic emoji boost done da!")
```
